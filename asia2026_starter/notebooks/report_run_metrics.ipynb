# Run Metrics Report

Loads all baseline/TabPFN run metrics into a DataFrame and prints all rows.


```python
from pathlib import Path
import json
import os
import pandas as pd

RUNS_ROOT = Path(os.environ.get('ASIA2026_RUNS_DIR', '/cluster/scratch/ppurkayastha/Acads/asia2026/runs'))

METRIC_KEYS = [
    'r2_sensory',
    'rmse_sensory',
    'mae_sensory',
    'r2_all',
    'rmse_all',
    'mae_all',
]

def extract_metrics(data):
    if isinstance(data, dict):
        if 'overall' in data and isinstance(data['overall'], dict):
            return data['overall']
        if 'cv_overall' in data and isinstance(data['cv_overall'], dict):
            return data['cv_overall']
    return data

rows = []
if RUNS_ROOT.exists():
    for run_dir in sorted(RUNS_ROOT.iterdir()):
        if not run_dir.is_dir():
            continue
        name = run_dir.name.lower()
        if 'baseline' not in name and 'tabpfn' not in name and 'pfn' not in name:
            continue
        for fname in ('cv_metrics.json', 'run_summary.json'):
            fpath = run_dir / fname
            if not fpath.exists():
                continue
            with fpath.open('r', encoding='utf-8') as f:
                data = json.load(f)
            metrics = extract_metrics(data)
            row = {
                'run_dir': run_dir.name,
                'metrics_file': fname,
            }
            for k in METRIC_KEYS:
                row[k] = metrics.get(k)
            rows.append(row)

df = pd.DataFrame(rows)
df
```

```python
# Print all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df)
```
